# Align plaque geojson for import in Xenium Explorer

```python
import geopandas as gpd
import geojson
import shapely
from shapely.ops import transform

df = gpd.read_file(r'D:/Xenium/H&E_staining/run3-aligment/3161-1_abeta_rotated.geojson')

```

```python
def swap_coords(x, y):
    return (20460-y, x)
```

```python
df2 = df.copy()
df2['geometry'] = df2.geometry.apply(lambda geom: transform(swap_coords, geom))
df3 = df2.copy()

# df3['geometry'] = df3.geometry.apply(lambda geom: geom.scale(-1,-1)
```

```python
# Save the GeoJSON FeatureCollection to a file
with open(f'D:/Xenium/H&E_staining/run3-aligment/3161-1_abeta_rotated_90_scale.geojson', 'w') as f:
    geojson.dump(df2, f)

print("GeoJSON saved")
```
